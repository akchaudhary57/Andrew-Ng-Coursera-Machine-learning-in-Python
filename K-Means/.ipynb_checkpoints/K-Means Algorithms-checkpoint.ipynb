## Write Machine learning equations for K-Means Clustering from Scratch (Andrew Ng Coursera)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from KMeans import *
```

```python
mat = loadmat("datasets/ex7data2.mat")
X = mat["X"]
```

```python
plt.plot(X[:,0],X[:,1],'yo')
```

## Cluster Assignment
![title](Figures/FindClosetdistance.PNG)

```python
kmeans = KMeans(3)
idx = kmeans.findClosestCentroids(X,np.array([[3,3],[6,2],[8,5]]))
print("Closest centroids for the first 3 examples:\n",idx[0:3])
```

## Move Centroid

![title](Figures/CentroidMeans.PNG)

```python
centriods = kmeans.computeCentroids(X,idx)
```

## Visualizing Kmeans 

```python
kmeans.plot_KMeans(X,idx,centriods,10)
```

## Random initialization

```python
initial_centriod = kmeans.kMeansInitCentroids(X)
idx = kmeans.findClosestCentroids(X,initial_centriod)
centriods = kmeans.computeCentroids(X,idx)
kmeans.plot_KMeans(X,idx,centriods,10)
```
