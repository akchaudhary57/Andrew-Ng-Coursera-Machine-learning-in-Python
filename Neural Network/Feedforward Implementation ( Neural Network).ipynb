## Andrew Ng Coursera dataset

```python
mat = scipy.io.loadmat('ex3data1.mat')
```

```python
X = mat['X']
y = mat['y'].ravel()
```

```python
n = 100 
index = np.random.choice(X.shape[0], n, replace=False) 
sel = X[index,:]
```

```python
regOneVsall = OneVsall()
regOneVsall.displaydata(sel)
```

```python
Weights = scipy.io.loadmat('ex3weights.mat')
```

```python
theta1 = Weights.get("Theta1")
theta2 = Weights.get("Theta2")
theta2 = np.roll(theta2, 1, axis=0)
```

```python
m = X.shape[0]
num_labels = Weights['Theta2'].shape[0]
```

```python
X = regOneVsall.append_one(X)
```

```python
FF = FeedForward_NN()
ztheta = FF.FeedForward(X,theta1,theta2)
```

### Prediction

```python
ypred = np.argmax(ztheta,axis=1)
ypred = [e if e else 10 for e in ypred]
```

```python
regOneVsall.score(y.flatten(),ypred)
```
